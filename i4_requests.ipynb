```python
from rdflib import Graph

import pandas as pd
import requests
```

```python
# This is sourced from the RDF tutorial
# Pros: No fixed schema -- established based on schema, can query any variety of data sets in the wikidata knowledge base.
# Cons: Difficult to learn and interpret as a beginner. This is a very limited tutorial set, since I need to develop more fluency with this form. The volume of domains to query means, that there are many examples but not all are easy to adapt to a unique usage.

# Create a Graph, parse in Internet data
g = Graph().parse("http://www.w3.org/People/Berners-Lee/card")

# Query the data in g using SPARQL
# This query returns the 'name' of all `foaf:Person` instances
q = """
    PREFIX foaf: <http://xmlns.com/foaf/0.1/>

    SELECT ?name
    WHERE {
        ?p rdf:type foaf:Person .

        ?p foaf:name ?name .
    }
"""

# Apply the query to the graph and iterate through results
for r in g.query(q):

    print(r["name"])
```
